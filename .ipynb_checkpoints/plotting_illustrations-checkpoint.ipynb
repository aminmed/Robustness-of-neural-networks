```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d
import os 
```

```python
 [ f.path for f in os.scandir("./experiments/FGSM") if f.is_dir() ]
```

```python
def plot_acc_eps(path_to_folder = "./experiments/FGSM") : 
    
    eps = {}
    confidences = pd.DataFrame(columns = ['softmax', 'epsilon']) 
    std_done = False 

    subfolders = [ f.path for f in os.scandir(path_to_folder) if f.is_dir() ]
    for folder in subfolders : 
        
        y_adv = np.load(folder + '/y_preds_adv.npy')
        y_std = np.load(folder + '/y_preds.npy')
        targets = np.load(folder + '/y_GT.npy')
        conf_adv = np.exp(np.load(folder + '/adv_confidences.npy'))
        conf_std = np.exp(np.load(folder + '/init_confidences.npy'))
        epsilon = float(folder.split('_')[2]) 
        acc_adv = np.sum(y_adv == targets) / len(targets)
        mask_correctely_predicated = (y_std == targets ) 
        
        success_rate = np.sum(y_adv[mask_correctely_predicated] != y_std[mask_correctely_predicated])\
                                        /len(targets)
        eps[epsilon] = [acc_adv * 100., success_rate * 100.]
        
        conf_std = conf_std[mask_correctely_predicated]
        conf_adv = conf_adv[mask_correctely_predicated]
        mask_high_std_conf = conf_std > 0.7
        print(np.sum(mask_high_std_conf))
        conf_std = conf_std[mask_high_std_conf]
        conf_adv = conf_adv[mask_high_std_conf]        

        if not std_done : 
            std_done = True 
            conf = pd.DataFrame({
                'softmax' : pd.Series((conf_std)), 
                'epsilon'  : pd.Series(['standard'] * len(conf_std))
            })
            confidences = pd.concat((confidences, conf), axis = 0)

        conf = pd.DataFrame({
                'softmax' : pd.Series((conf_adv)), 
                'epsilon'  : pd.Series(['%.3f'%(epsilon/255.)] * len(conf_adv))
        })
        confidences = pd.concat((confidences, conf), axis = 0)
        
        
    x = np.array([e / 255. for e in eps.keys()])
    y = np.array([eps[e][0] for e in eps.keys()])
    z = np.array([eps[e][1] for e in eps.keys()])
    c1 = interp1d(x, y, kind = "cubic")
    c2 = interp1d(x, z, kind = "cubic")
    # Plotting the Graph
    X_=np.linspace(x.min(), x.max(), len(x))
    Y_=c1(X_)
    Z_=c2(X_)
    plt.plot(X_, Y_, label = "accuracy on adversarial images")
    plt.plot(X_, Z_, label = "attack success rate")
    
    plt.title("Accuracy and success rates of L2PGD attack")
    plt.xlabel("epsilon")
    plt.ylabel("accuracy %")
    plt.legend(loc = 'lower left')
    plt.savefig('L2PGD_adv_acc.png', transparent=True)
    plt.show()
    

    return eps, confidences

```

```python
_, c = plot_acc_eps("./experiments/L2PGD")
```

```python
c = c.sort_values(by = ['epsilon'])
```

```python
std_c = c['epsilon'] == 'standard' 
std_c = c[std_c]
std_c = std_c.reset_index() 
c = pd.concat((std_c, c), axis = 0)
```

```python
import seaborn as sns
sns.boxplot(data=c, y="softmax", x="epsilon")
plt.savefig('L2PGD_boxplots.png', transparent=True)
```

```python
path = './experiments/FGSM/FGSM_eps_3._time_20221208155706'
```

```python
import torchvision as tv 
import torchvision.transforms as transforms
from torch.utils.data import DataLoader
adverserial_images = np.load(path + '/adverserial_images.npy')
y_preds = np.load(path + '/y_preds.npy')
y_preds_adv = np.load(path + '/y_preds_adv.npy')
test_label = np.load(path + '/y_GT.npy')
ds = tv.datasets.CIFAR10("./data", 
                                       train=False, 
                                       transform=transforms.Compose([transforms.ToTensor()]), 
                                       download=True)

l = DataLoader( ds,
                            batch_size=128, 
                            shuffle=False, num_workers=4)
test_images = []
for images, labels in l : 
    test_images.extend(images.cpu().data.numpy())


# Get index of  all the images where the attack is succesful
attack = (y_preds != y_preds_adv)
indexes = []
for i in range(len(attack)):
  if attack[i] == True:
    indexes.append(i)

indexes = np.array(indexes)
```

```python
test_images =np.array(test_images)
```

```python
# Plot the images 
classes = ('plane', 'car', 'bird', 'cat', 'deer', 'dog', 'frog', 'horse', 'ship', 'truck')
plt_idx = 0
while plt_idx < 2:
    idx = np.random.choice(indexes)
    img = test_images[idx]
    adv_img = adverserial_images[idx]

    img = np.transpose(img,(1,2,0))
    img = img.clip(0,1)
   
    adv_img =np.transpose(adv_img,(1,2,0))
    adv_img = adv_img.clip(0,1)
    noise = adv_img - img
    noise = np.absolute(10*noise)  # Noise is multiplied by 10 for visualisation purpose
    noise = noise.clip(0,1)
    
    if y_preds[idx] != y_preds_adv[idx]:
        disp_im = np.concatenate((img, adv_img,noise), axis=1)
        ax = plt.subplot(1,2,plt_idx+1)
#         ax.set_title(classes[test_label[idx]],classes[y_preds[idx]], classes[y_preds_adv[idx]] )
        ax.set_title("pred: {}, adv:{}".format(y_preds[idx], y_preds_adv[idx]))
        plt.imshow(disp_im)
        plt.xticks([])
        plt.yticks([])
        plt_idx += 1
        print("True Label: ",classes[test_label[idx]]," ","Predicted Label:",classes[y_preds[idx]]," ", "Adverserial Label:",classes[y_preds_adv[idx]])
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

def img_is_color(img):

    if len(img.shape) == 3:
        # Check the color channels to see if they're all the same.
        c1, c2, c3 = img[:, : , 0], img[:, :, 1], img[:, :, 2]
        if (c1 == c2).all() and (c2 == c3).all():
            return True

    return False

def show_image_list(list_images, list_titles=None, list_cmaps=None, grid=False, num_cols=3, figsize=(3, 3), title_fontsize= 10):
    '''
    Shows a grid of images, where each image is a Numpy array. The images can be either
    RGB or grayscale.
    Parameters:
    ----------
    images: list
        List of the images to be displayed.
    list_titles: list or None
        Optional list of titles to be shown for each image.
    list_cmaps: list or None
        Optional list of cmap values for each image. If None, then cmap will be
        automatically inferred.
    grid: boolean
        If True, show a grid over each image
    num_cols: int
        Number of columns to show.
    figsize: tuple of width, height
        Value to be passed to pyplot.figure()
    title_fontsize: int
        Value to be passed to set_title().
    '''

    assert isinstance(list_images, list)
    assert len(list_images) > 0
    assert isinstance(list_images[0], np.ndarray)

    if list_titles is not None:
        assert isinstance(list_titles, list)
        assert len(list_images) == len(list_titles), '%d imgs != %d titles' % (len(list_images), len(list_titles))

    if list_cmaps is not None:
        assert isinstance(list_cmaps, list)
        assert len(list_images) == len(list_cmaps), '%d imgs != %d cmaps' % (len(list_images), len(list_cmaps))

    num_images  = len(list_images)
    num_cols    = min(num_images, num_cols)
    num_rows    = int(num_images / num_cols) + (1 if num_images % num_cols != 0 else 0)

    # Create a grid of subplots.
    fig, axes = plt.subplots(num_rows, num_cols, figsize=figsize)
    
    # Create list of axes for easy iteration.
    if isinstance(axes, np.ndarray):
        list_axes = list(axes.flat)
    else:
        list_axes = [axes]

    for i in range(num_images):

        img    = list_images[i]
        title  = list_titles[i] if list_titles is not None else 'Image %d' % (i)
        cmap   = list_cmaps[i] if list_cmaps is not None else (None if img_is_color(img) else 'gray')
        
        list_axes[i].imshow(img, cmap=cmap)
        list_axes[i].set_title(title, fontsize=title_fontsize) 
        list_axes[i].grid(grid)
        list_axes[i].axis('off')
    for i in range(num_images, len(list_axes)):
        list_axes[i].set_visible(False)

    fig.tight_layout()
    plt.savefig('example.png', format='png', transparent=True)
    _ = plt.show()
```

```python
idx = np.random.choice(indexes)
img = test_images[idx]
adv_img = adverserial_images[idx]
img = np.transpose(img,(1,2,0))
img = img.clip(0,1)   
adv_img =np.transpose(adv_img,(1,2,0))
adv_img = adv_img.clip(0,1)
noise = adv_img - img
noise = np.absolute(10*noise)  # Noise is multiplied by 10 for visualisation purpose
noise = noise.clip(0,1)

show_image_list(list_images =[img, adv_img] , list_titles=[classes[y_preds[idx]], classes[y_preds_adv[idx]]])
```
